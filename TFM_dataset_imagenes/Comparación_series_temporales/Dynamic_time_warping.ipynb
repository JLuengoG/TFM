# DTW

```python
# Prueba para ver como funciona DTW
###################################
# Cada fila es una serie temporal comparada con las demas
# Al añadir una serie random al final es la última fila de la matriz

random_array = np.random.rand(1, 1600)
prueba = np.concatenate((x_train_label1_20filas[:5], random_array))

# matriz de distancias
matrix = dtw.distance_matrix_fast(prueba)

# Imprimir matriz de 14x14
for i in matrix:
    print('\t'.join(map(str, i)))
```

# Comparacion dentro de la misma etiqueta

```python
 def dtw_media(data, label):
  # matriz de distancias
  matrix = dtw.distance_matrix_fast(data)

  # # Imprimir matriz
  # for i in matrix:
  #     print('\t'.join(map(str, i)))

  # Convertir a DataFrame
  df_matrix =pd.DataFrame(matrix)

  print(f'***{label}***')

  # Media global entre las series
  mean_total = df_matrix.mean(axis=1).mean()
  print(f'La media de cada serie temporal con las demas es: {mean_total} \n')


  # Calcular la media de cada serie con las demas
  mean_filas = df_matrix.mean(axis=1)
  print(f'La media de cada serie temporal con las demas es: \n{mean_filas} \n')
```

```python
dtw_media(x_train_label1_20filas, 'label1')
dtw_media(x_train_label2_20filas, 'label2')
dtw_media(x_train_label3_20filas, 'label3')
dtw_media(x_train_label4_20filas, 'label4')
dtw_media(x_train_label5_20filas, 'label5')
dtw_media(x_train_label6_20filas, 'label6')
dtw_media(x_train_label7_20filas, 'label7')
dtw_media(x_train_label8_20filas, 'label8')
dtw_media(x_train_label9_20filas, 'label9')
```

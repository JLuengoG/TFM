# Graficar juntos

```python
def graficar_juntas(data):
  # Graficar el espectro de frecuencia
  plt.figure(figsize=(20, 5))
  for fila in data:
    plt.plot(list(range(0, len(fila))), fila)
  plt.xlabel("Muestra")
  plt.ylabel("Normalizada")
  plt.grid()
  plt.show()
```

```python
graficar_juntas(df_x_train_norm_label0_200filas)
graficar_juntas(df_x_train_norm_label1_50filas)
graficar_juntas(df_x_train_norm_label2_50filas)
graficar_juntas(df_x_train_norm_label3_50filas)
graficar_juntas(df_x_train_norm_label4_50filas)
graficar_juntas(df_x_train_norm_label5_50filas)
graficar_juntas(df_x_train_norm_label6_50filas)
graficar_juntas(df_x_train_norm_label7_50filas)
graficar_juntas(df_x_train_norm_label8_50filas)
graficar_juntas(df_x_train_norm_label9_50filas)
```

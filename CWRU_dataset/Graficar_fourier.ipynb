# Graficar juntos

```python
def graficar_juntas_fourier(data):
  # Graficar el espectro de frecuencia
  plt.figure(figsize=(20, 5))
  for fila in data:
    plt.plot(frecuencias, fila)
  plt.xlabel("Frecuencia Hz")
  plt.ylabel("Amplitud normalizada")
  plt.grid()
  plt.show()
```

```python
graficar_juntas_fourier(df_x_train_fourier_norm_label0_200filas)
graficar_juntas_fourier(df_x_train_fourier_norm_label1_50filas)
graficar_juntas_fourier(df_x_train_fourier_norm_label2_50filas)
graficar_juntas_fourier(df_x_train_fourier_norm_label3_50filas)
graficar_juntas_fourier(df_x_train_fourier_norm_label4_50filas)
graficar_juntas_fourier(df_x_train_fourier_norm_label5_50filas)
graficar_juntas_fourier(df_x_train_fourier_norm_label6_50filas)
graficar_juntas_fourier(df_x_train_fourier_norm_label7_50filas)
graficar_juntas_fourier(df_x_train_fourier_norm_label8_50filas)
graficar_juntas_fourier(df_x_train_fourier_norm_label9_50filas)
```

# Import

```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.fft import fft, fftfreq
```

# Cargar datos

```python
x_train = np.load('/content/drive/MyDrive/TFM/Google_Colab/CWRU/CRWU_Javi/x_train.npy')
```

# Transformar a fourier

```python
frecuencia_muestreo = 48000

# Realizar la transformada de Fourier utilizando scipy
fft_result = fft(x_train[0])
frecuencias = fftfreq(len(fft_result), 1/frecuencia_muestreo)

# Graficar el espectro de frecuencia
plt.figure(figsize=(20, 5))
plt.plot(frecuencias, np.abs(fft_result))
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Amplitud")
plt.title("Espectro de Frecuencia")
plt.grid()
plt.show()
```

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

# Original data

```python
data_npy = np.load(f'./CWRU_Original_data/signal_data.npy')
labels_npy = np.load(f'./CWRU_Original_data/signal_data_labels.npy')
```

```python
df = pd.DataFrame(data_npy)
```

### Create labels column

```python
# Create array with labels
np_labels = []
n_rows = 280
for label_id in list(range(0,10)):
    np_label_id = np.full(n_rows, int(label_id))
    np_labels = np.concatenate((np_labels, np_label_id), axis=None)
    
# Create column
df.insert(0, "label_id", np_labels)
```

# Plot labels

```python
def plot(df, label):
    title = f'Label {label}'
    df = df.T
    df.plot(title= title, figsize=(20, 5), legend=False, colormap= 'Blues')
      
```

```python
for label, data in df.groupby('label_id'):
    data.drop("label_id", axis=1, inplace=True)
    # Plot first 10 time series
    plot(data.head(5), label)
```
